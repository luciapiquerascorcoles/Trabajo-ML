# 1. Dummies y escalado

Vimos que obteníamos demasiadas variables. La __variable `addr_state`__tiene demasiados tipos (muchos Estados), por tanto, decidimos __quitarla__ y así poder conseguir un número menor de variables. Además, hay que decir que cuando hicimos el Análisis Exploratorio, tampoco había muchas diferencias entre individuos que pagarón o cancelaron según el Estado en el que vivían. Nos quedamos así con 19 variables.

```python
# Eliminamos la variable de 'addr_state'
LOAN.drop("addr_state", axis=1, inplace=True)
```

```python
LOAN
```

__Diferenciamos__ en dos objetos las __variables numéricas y las categóricas__:

```python
variables_numericas = LOAN.select_dtypes(include = [np.number])
variables_categoricas = LOAN.select_dtypes(include = [np.object])
```

```python
variables_categoricas
```

```python
# Vemos las categóricas (8):
variables_categoricas.columns
```

```python
# Vemos las numéricas (11):
variables_numericas.columns
```

## 1.1 Normalizamos las variables numéricas

```python
scaler = preprocessing.StandardScaler()

numericas_escaladas = scaler.fit_transform(variables_numericas)
numericas_escaladas = pd.DataFrame(numericas_escaladas, columns = [variables_numericas])
```

Vemos como han quedado las variables después de normalizarlas:

```python
numericas_escaladas.head()
```

## 1.2 Convertimos la variable _loan_status_ en _dummie_ 

```python
variables_categoricas['Charged_off'] = (variables_categoricas['loan_status'] == 'Charged Off').apply(np.uint8)
variables_categoricas.drop('loan_status', axis=1, inplace=True)
```

## 1.3 Dummies del resto de variables categóricas

Los algoritmos de aprendizaje automático requieren que las variables de entrada y salida sean numéricas. Esto implica que los datos categóricos deben ser codificados en números antes de que podamos usarlos para ajustar y evaluar un modelo. Hay muchas maneras de codificar variables categóricas para el modelado.

Utilizamos el método de _One-Hot Encoding_ con la función `pd.get_dummies`.

```python
variables_categoricas = pd.get_dummies(variables_categoricas, columns=['application_type', 'home_ownership', 'initial_list_status', 'purpose',
       'sub_grade', 'term', 'verification_status'])


```

```python
variables_categoricas
```

```python
variables_categoricas.index = range(variables_categoricas.shape[0])
```

```python
variables_categoricas.dtypes
```

Concatenamos ambos dataframes:

```python
LOAN_var_norm = pd.concat([numericas_escaladas, variables_categoricas], axis = 1)
```

```python
LOAN_var_norm
```

```python
LOAN_var_norm.shape
```

El dataframe resultante se queda en 73 variables. 

```python
# Guardamos el dataframe con el que entrenaremos los modelos:
LOAN_var_norm.to_csv('../data/intermediate/LOAN_var_norm.csv', index=False)
```
